# Prediction using Unsupervised ML (Level - Beginner)

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
# Loading the dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
# Assining the attributes as Input variable and Target variable
X = iris.data[:, :2]    # For only Sepal length and width 
y = iris.target         # Spieces
```

```python
# Scatter plot
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Spea1 Length', fontsize=15)
plt.ylabel('Sepal Width', fontsize=15)
```

# K-Means Clustering

```python
# To find the required number of clusters for K-means clustering

from sklearn.cluster import KMeans
css = []    # Cluster sum of squares

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    
    css.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, allowing us to observe 'The elbow' in the gragh
plt.plot(range(1, 11), css)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('CSS') 
plt.grid()
plt.show()
```

**From the above elbow graph, the optimum clusters is where the elbow occurs. Here is where the within cluster sum of squares (CSS) doesn't decrease significantly with every iteration.**

**From this we choose the number of clusters as ** '3**'.

```python
from sklearn.cluster import KMeans

# Creating an instance
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)

#Fitting the model
km.fit(X)
```

```python
# To Identify the center points of the data
centers = km.cluster_centers_
print(centers)
```

```python
# The lowest SSE value (Error Sum of Squares)
kmeans.inertia_
```

```python
# The number of iterations required to converge
kmeans.n_iter_
```

**Compare the original data versus and the clustered results, this will tell us to which cluster does the data observations belong**

```python
new_labels = km.labels_

# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(14,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='cool',
edgecolor='k', s=150)

# Plotting the centroids of the clusters
axes[0].scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1], 
            s = 150, c = 'yellow', label = 'Centroids')
axes[1].scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1], 
            s = 150, c = 'yellow', label = 'Centroids')

axes[0].set_xlabel('Sepal length', fontsize=15)
axes[0].set_ylabel('Sepal width', fontsize=15)
axes[1].set_xlabel('Sepal length', fontsize=15)
axes[1].set_ylabel('Sepal width', fontsize=15)
axes[0].tick_params(direction='in', length=10, width=8, colors='k', labelsize=12)
axes[1].tick_params(direction='in', length=10, width=8, colors='k', labelsize=12)
axes[0].set_title('Actual', fontsize=15)
axes[1].set_title('Predicted', fontsize=15)
```

**This Compares the K-Means Clustering for Actual and Predicted values**
